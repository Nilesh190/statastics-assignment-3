## Question 1: What is hypothesis testing in statistics?
**Answer:**
Hypothesis testing is a statistical method that helps in making decisions or inferences about population parameters based on sample data. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), collecting sample data, and then using statistical tests to determine whether to reject the null hypothesis.

## Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
**Answer:**
The null hypothesis (H0) is a statement that there is no effect or no difference in the population. It is the hypothesis that we aim to test. The alternative hypothesis (H1) is the statement that indicates the presence of an effect or difference. If the evidence from sample data is strong enough, the null hypothesis is rejected in favor of the alternative hypothesis.

## Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
**Answer:**
The significance level (denoted as alpha, α) is the probability threshold used to decide whether to reject the null hypothesis. Common values are 0.05 or 0.01. If the p-value obtained from the test is less than α, the null hypothesis is rejected, indicating that the observed effect is statistically significant.

## Question 4: What are Type I and Type II errors? Give examples of each.
**Answer:**
- **Type I error:** Rejecting the null hypothesis when it is actually true. Example: Concluding a new drug is effective when it is not.
- **Type II error:** Failing to reject the null hypothesis when it is actually false. Example: Concluding a new drug is ineffective when it actually is effective.

## Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
**Answer:**
- **Z-test:** Used when the population variance is known and the sample size is large (usually n > 30).
- **T-test:** Used when the population variance is unknown and the sample size is small (n ≤ 30). It accounts for additional uncertainty in the estimation of the standard deviation.

## Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate binomial distribution data
n = 10
p = 0.5
size = 1000
data = np.random.binomial(n, p, size)

# Plot the histogram
plt.hist(data, bins=range(n+2), edgecolor='black', density=True)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

## Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis testing parameters
mu = 50  # Population mean

# Calculate sample mean and standard error
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)
standard_error = sample_std / np.sqrt(n)

# Compute Z statistic
z_score = (sample_mean - mu) / standard_error

# Calculate p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f'Sample Mean = {sample_mean}')
print(f'Z-Score = {z_score}')
print(f'P-Value = {p_value}')

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

## Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate normal distribution data
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=100)

# Calculate mean and 95% confidence interval
mean = np.mean(data)
std_err = np.std(data, ddof=1) / np.sqrt(len(data))
confidence_interval = stats.norm.interval(0.95, loc=mean, scale=std_err)

print(f'Mean = {mean}')
print(f'95% Confidence Interval = {confidence_interval}')

# Plot the data
plt.hist(data, bins=15, edgecolor='black', density=True)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean = {mean:.2f}')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.random.normal(50, 5, 100)

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    return (data - mean) / std

z_scores = calculate_z_scores(data)

print(f'Z-Scores:\n{z_scores[:10]}')  # Display first 10 for brevity

# Plot histogram
plt.hist(z_scores, bins=15, edgecolor='black', density=True)
plt.title('Histogram of Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.show()

# Explanation
print("Z-scores represent how many standard deviations a data point is from the mean. A Z-score of 0 means the point is exactly at the mean.")
```
